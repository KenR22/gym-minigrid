```python
from gym_minigrid.minigrid import *
from gym_minigrid.register import register
```

```python
class InnateBehaviorWrapper(gym.Wrapper):
    """
    Changes the state of Agent automatically if lava is within the view of the agent
    """

    def __init__(self, env):
        self.skip_counter=0
        super().__init__(env)

    def step(self, action):
        obs, reward, done, info = self.env.step(action)

        # Tuple based on which we index the counts
        # We use the position after an update
        
        #print(np.any(self.env.gen_obs()["image"]==9))
        if (np.any(obs["image"]==9)):
            self.agent_state=-1
            self.skip_counter=5
        else:
            self.agent_state=1
            self.skip_counter-=1
        obs = self.env.gen_obs()

        return obs, reward, done, info

```

```python
env = InnateBehaviorWrapper(gym.make('MiniGrid-LavaGapS7-v0'))
```

```python
plt.imshow(env.render('rgb_array'))
observation, reward, done, info = env.step(1)
print(env.skip_counter,env.agent_state)
```
